## 裝飾器 (Decorator)

### Return Function

```python
def main():
  print('main')

def decorator(func):

  def wrap_func():
    print('Before')
    func()
    print('After')

  return wrap_func


main()
```

```python
f = decorator(main)
f()
```

### Decorator

```python
def decorator(func):

  def wrap_func():
    print('Before')
    func()
    print('After')

  return wrap_func

@decorator
def main():
  print('main')

main()
```
