### 覆寫(Override)特殊方法

```python
class Wizard(object):
  def __init__(self, name):
    self.name = name
a = Wizard('irocks')
```

```python
dir(a)
```

```python
print(a)
```

```python
class Wizard(object):
  def __init__(self, name):
    self.name = name
  def __str__(self):
    return 'i am the mighty wizard, my name is {}'.format(self.name)
    
a = Wizard('irocks')
```

```python
print(a)
```

```python
del a
```

```python
print(a)
```

```python
class Wizard(object):
  def __init__(self, name):
    self.name = name
  def __str__(self):
    return 'i am the mighty wizard, my name is {}'.format(self.name)
  def __del__(self):
    print('Destroy the wizard')
    
a = Wizard('irocks')
del a
```
