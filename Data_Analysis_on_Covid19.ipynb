## Covid19 資料科學分析

### 資料載入

```python
import pandas
df = pandas.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
```

```python
df.head()
```

### 資料前處理

```python
df.set_index('Country/Region', inplace=True)
df.head()
```

```python
df = df.iloc[:,3:]
```

```python
df.head()
```

```python
df = df.T
```

```python
df.head()
```

```python
df['US']
```

### 資料視覺化

```python
df['US'].plot(kind = 'line', figsize = [20,10])
```

```python
df[['US','India']].plot(kind = 'line', figsize = [20,10])
```

### 統計分析

```python
df.iloc[-1,:].sort_values(ascending = False).head(10)
```

```python
df.iloc[-1,:].sort_values(ascending = False).head(10).plot(kind = 'bar')
```
