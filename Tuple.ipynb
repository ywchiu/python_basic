### Tuple 基本操作

```python
t1 = (1,2,3)
t1
```

```python
type(t1)
```

```python
t2 = 1,2,3
t2
```

```python
l1 = [1,2,3]
t3 = tuple(l1)
t3
```

```python
t1[0]
```

```python
len(t1)
```

```python
t1[-1]
```

```python
t1.append(4)
```

```python
t1[1] = 200
```

### Tuple 與 List 差別

```python
l1 = [1,2,3]
print(dir(l1))
```

```python
t1 = (1,2,3)
print(dir(t1))
```

```python
t1 = (1,1,2,2,2,3)
t1.count(2)
```

```python
t1.index(2)
```
