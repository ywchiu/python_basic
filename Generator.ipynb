## 生成器 (Generator)

### 建立生成器

```python
def gencube(num):
  ary = []
  for i in range(num):
    ary.append(i ** 3)
  return ary

gencube(10)
```

```python
for i in gencube(10):
  print(i)
```

```python
def gencube2(num):
  for i in range(num):
    yield i ** 3

gencube2(10)
```

```python
for i in gencube2(10):
  print(i)
```

### 生成器 (Generator) 與一般函式 (Normal Function)


```python
def fibonacci(num):
  a,b = 1,1
  output = []
  for i in range(num):
    output.append(a)
    a,b = b,a+b
  return output

fibonacci(10)
```

```python
def genfibonacci(num):
  a,b = 1,1
  for i in range(num):
    yield a
    a,b = b,a+b

genfibonacci(10)
```

```python
for i in genfibonacci(10):
  print(i)
```
