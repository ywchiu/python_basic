## While 迴圈 (While Loop)

### While 迴圈操作

```python
s = 0
while s < 10:
  print(s)
  s = s + 1 
```

### While...Else

```python
s = 0
while s < 10:
  print(s)
  s = s + 1 
else:
  print('Executed Successfully')
```

```python
s = 0
while s < 10:
  print(s)
  s = s + 1 
  break
else:
  print('Executed Successfully')
```

### Continue, Pass, Break

```python
s = 0
while s < 10:
  print(s)
  s = s + 1 
  # break
  if s == 3:
    break
```

```python
s = 0
while s < 10:
  print(s)
  s = s + 1 
  # pass
  if s == 3:
    pass
```

```python
s = 0
while s < 10:
  print(s)
  s = s + 1 
  # pass
  if s == 3:
    pass
    print('find my lucky number')
```

```python
s = 0
while s < 10:
  print(s)
  s = s + 1 
  # continue
  if s == 3:
    continue
    print('find my lucky number')
```

### 無窮迴圈 (Infinite Loop)

```python
s = 0 
while s < 10:
  print(s)
```

### 練習題

我們希望建立一個對話機器人, 機器人裡面內建一個資料問答資料庫, 資料庫的問題跟答案如下所示:

- 問題:保險 答案: 以下是我們推薦的險總類
- 問題:貸款 答案: 以下是我們推薦的貸款總類
- 問題:存款 答案: 以下是您的銀行存款餘額

希望你能用這個問答資料庫建立成一個Python 字典, 當使用者輸入問題時, 只要問題中有部分關鍵字對到問答資料庫的問題,電腦就予以回答

但如果使用者如果今天沒有提到「**離開對話**」，則機器人會持續問下個問題

```python
qa = {'保險': '以下是我們推薦的險總類', '貸款': '以下是我們推薦的貸款總類', '存款': '以下是您的銀行存款餘額'}
question = input('請輸入您的問題:')
#print(question)

while question != '離開對話':
  for q in qa:
    if q in question:
      print(qa.get(q))
      break
  question = input('請輸入您的問題:')
```
