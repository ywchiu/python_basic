## 巢狀陳述與範圍 (Nested Statement and Scope)

### Local

```python
f = lambda x : x ** 2
```

### Enclosing Functions

```python
name = 'David'

def hello():
  name = 'John'
  def greet():
    print(name)
  greet()

hello()
```

### Global

```python
print(name)
```

### Built-in

```python
len
```

### Local Variable v.s. Global Variable

```python
x = 5

def getvalue():
  x = 2 
  print(x)

getvalue()
```

```python
print(x)
```

```python
x = 5

def getvalue():
  global x
  x = 2 
  print(x)

getvalue()
```

```python
print(x)
```
