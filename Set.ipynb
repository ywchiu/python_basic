### 將 List 轉換成 Set

```python
li = [1,1,2,2,2,3,3]
set(li)
```

### Set 基礎操作

```python
A = [1,2,3,1,3,4,3]
B = [2,4,5,6,5]
```

```python
set(A) & set(B)
```

```python
set(A) | set(B)
```

```python
set(A) - set(B)
```

```python
set(B) - set(A)
```
