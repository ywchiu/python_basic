## 錯誤與例外

### Try & Except

`````python
divisor = input('Divisor:')
dividend = input('Dividend:')
try:
  print(float(dividend) / float(divisor))
except:
  print('ERROR')
`````

### Exception

`````python
divisor = input('Divisor:')
dividend = input('Dividend:')
try:
  print(float(dividend) / float(divisor))
except Exception as e:
  print('ERROR', e)
`````

### Finally

`````python
divisor = input('Divisor:')
dividend = input('Dividend:')
try:
  print(float(dividend) / float(divisor))
except Exception as e:
  print('ERROR', e)
finally:
  print('WHATEVER THE RESULT IS, PRINT THIS')
`````

### Else

`````python
divisor = input('Divisor:')
dividend = input('Dividend:')
try:
  print(float(dividend) / float(divisor))
except Exception as e:
  print('ERROR', e)
else:
  print('IF SUCCESS, PRINT THIS')
finally:
  print('WHATEVER THE RESULT IS, PRINT THIS')
`````

### 練習題
有一個客戶存款餘額有：300000，現在請你寫一個程式，使用者可以輸入(input)客戶想要提領(withdraw)的金額，如果客戶的餘額夠提領，輸入完後，螢幕上會秀出以下字串:

````
您已經提領了 <withdraw> 元整，您的餘額還剩下 <deposit> 元整
````

但如果客戶的餘額不夠提領，輸入完後，螢幕上會秀出以下字串:
**您的餘額不足**

但當如果使用者輸入的資料不為數字，螢幕上會秀出： **請輸入數字**

`````python
deposit = 300000
withdraw = input('請輸入您要提領的金額:')
try:
  deposit = deposit - int(withdraw)

  if deposit >= 0:
    print(f'您已經提領了 {withdraw} 元整，您的餘額還剩下 {deposit} 元整')
  else:
    print('您的餘額不足')
except:
  print('請輸入數字')
`````
