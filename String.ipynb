## 練習題

現在客戶要將 10,000 台幣換成美金， 假設：

- 原始輸入字串為: NTD$ 10,000 
- 匯率為 : 27.3 

請將該金額轉換成美金整數金額

**提示:使用replace 與 int 作轉換**

```python
s = 'NTD$ 10,000'
?s.replace
```

```python
t1 = s.replace('NTD$ ', '')
t2 = t1.replace(',', '')
t3 = int(t2)
t4 = t3 / 27.3
result = int(t4)
result
```

```python
val = int(s.replace('NTD$ ', '').replace(',', ''))
int(val / 27.3)
```
