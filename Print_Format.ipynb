## 輸出格式

```python
name = 'David Chiu'
position = '主持人'
company = '大數軟體'

'大家好！我的名字是' + name + '，我在' + company + '擔任' + position
```

### 使用 %s 控制輸出格式

```python
s = '大家好！我的名字是%s，我在%s擔任%s'
s%(name, company, position)
```

### 使用 {} 控制輸出格式

```python
s = '大家好！我的名字是{}，我在{}擔任{}'
s.format(name, company, position)
```

```python
s = '大家好！我的名字是{0}，我在{1}擔任{2}'
s.format(name, company, position)
```

```python
s = '大家好！我的名字是{n}，我在{c}擔任{p}'
s.format(n = name, c = company, p = position)
```

### 使用 f-string 控制輸出格式

```python
name = 'David Chiu'
position = '主持人'
company = '大數軟體'
s = f'大家好！我的名字是{name}，我在{company}擔任{position}'
s
```
