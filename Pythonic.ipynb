### 真假比較

```python
name = 'David'
langs = ['python', 'R', 'julia']
info = {'name':'David', 'Sex':'Male', 'age':18}
```

```python
# Non-Pythonic
if name != '' and len(langs) > 0 and info != {}:
    print('All True')
```

```python
# Pythonic
if name and langs and info:
  print('All True')
```

### 字串反轉

```python
s = '上海自來水來自海上'
```

```python
# Non-Pythonic
def reverse_str(s):
    t = ''
    for x in range(len(s)-1, -1,-1):
        #print(x)
        t += s[x]
    return(t)

reverse_str(s)
```

```python
# Pythonic
s[::-1]
```

```python
reverse_str = lambda s : s[::-1]
reverse_str(s)
```

```python
# Pythonic
' '.join(strList)
```

### 求和與乘積

```python
numList = [1,2,3,4,5]  
```

```python
# Non-Pythonic
s = 0
maxNum = -float('inf')
minNum = float('inf')
prod = 1
for num in numList:
    if num > maxNum:
        maxNum = num
    if num < minNum:
        minNum = num
    s += num
    prod *= num
prod
print(s, maxNum, minNum, prod)
```

```python
# Pythonic
sum(numList)
```

```python
max(numList)
```

```python
min(numList)
```

```python
from operator import mul
from functools import reduce
reduce(mul, numList)
```

### 資料篩選

```python
# Non-Pythonic
l = []
for x in range(10):
    if x % 3 == 0:
        l.append(x*x)
l
```

```python
# Pythonic
[x*x for x in range(10) if x % 3 == 0]
```

### 字典預設值

```python
items = ['apple', 'apple', 'apple', 'book', 'book', 'cat']
```

```python
# Non-Pythonic
dic = {}
for item in items:
  if item not in dic:
    dic[item] = 1
  else:
    dic[item] = dic[item] + 1
dic
```

```python
# Pythonic
dic = {}
for item in items:
  dic[item] = dic.get(item, 0) + 1
dic
```

### For … Else

```python
# Non-Pythonic
find = False
for x in range(1,5):
    if x == 5:
        find = True
        print('find 5')
        break
if not find:
    print('can not find 5!')
```

```python
# Pythonic
for x in range(1,5):
    if x == 5:
        print('find 5')
        break
else:
    print('can not find 5!')
```

### 條件判斷式

```python
a = 3   
```

```python
# Non-Pythonic
if a > 2:
    b = 2
else:
    b = 1
```

```python
# Pythonic
b = 2 if a > 2 else 1
```

### Enumerate

```python
array = [1, 2, 3, 4, 5]
```

```python
# Non-Pythonic
for i in range(len(array)):
    print(i, array[i])
```

```python
# Pythonic
for idx, ele in enumerate(array):
  print(idx, ele)
```

### ZIP

```python
keys = ['David', 'Marry', 'Joseph']
values = [50000, 70000, 38000]
```

```python
# Non-Pythonic
dic = {}
for i,e in enumerate(keys):
    dic[e] = values[i]
dic
```

```python
# Pythonic
dic = {}
for idx, ele in zip(keys, values):
  dic[idx] = ele
dic
```
