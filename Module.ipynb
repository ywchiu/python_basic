## 模組 (Module)

### 內建模組 (Built-in Module)

```python
a = 3.8

import math
math.ceil(a)
```

### 檢視模組 (dir)

```python
dir(math)
```

### 使用者定義模組 (User-Defined Module)

```python
import data_loading
print(data_loading.a)
```

```python
data_loading.func('David')
```

### 模組名稱空間 (Module Namespace) 

```python
import data_loading
data_loading.func('David')
```

```python
from data_loading import func
func('David')
```

```python
from data_loading import func as f
f('David')
```

```python
from data_loading import *
func('David')
```

```python
def func(n):
  print('Bye bye,' + n)
  
func('David')
```

```python
from data_loading import func
func('David')
```
