### IF...ELSE Statement

`````python
a = 5
b = 7
if a > b :
  print('a 比 b 來得大')
else:
  print('a 不比 b 來得大')
`````

`````python
# Pythonic
a = 7 
b = 5
'a 比 b 來得大' if a > b else 'a 不比 b 來得大'
`````

### IF...ELIF...ELSE 陳述式 (Statement)

`````python
deposit = 15000

if deposit > 30000:
  print('買 Iphone')
elif deposit > 20000:
  print('買 小米')
elif deposit > 10000:
  print('買 白牌手機')
else:
  print('我要存更多錢')
`````

### Switch...Case 陳述式 (Statement)

`````python
q = '理財'
if q == '貸款':
  print('請洽貸款專員')
elif q == '理財':
  print('請洽理財專員')
else:
  print('請洽總機')
`````

`````python
q = '請問最近的分行在哪?'
{'貸款': '請洽貸款專員', '理財': '請洽理財專員'}.get(q, '請洽總機')
`````

### 練習題
有一個客戶存款餘額有：300000，現在請你寫一個程式，使用者可以輸入(input)客戶想要提領(withdraw)的金額，如果客戶的餘額夠提領，輸入完後，螢幕上會秀出以下字串:

````
您已經提領了 <withdraw> 元整，您的餘額還剩下 <deposit> 元整
````

但如果客戶的餘額不夠提領，輸入完後，螢幕上會秀出以下字串:
您的餘額不足


`````python
deposit = 300000
withdraw = input('請輸入您要提領的金額:')
deposit = deposit - int(withdraw)
if deposit >= 0:
  print(f'您已經提領了 {withdraw} 元整，您的餘額還剩下 {deposit} 元整')
else:
  print('您的餘額不足')
`````
