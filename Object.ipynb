### 方法 (Method)

```python
class Wizard(object):
  professions = 'Wizard'
  def __init__(self, name, server):
    self.name = name
    self.server = server
  def walk(self):
    print('walking')
  def rest(self):
    print('resting')
  def run(self):
    print('running')
    
a = Wizard('irocks', 'A')
```

```python
a.run()
```

```python
a.walk()
```

```python
class Wizard(object):
  professions = 'Wizard'
  def __init__(self, name, server):
    self.name = name
    self.server = server
    self.hp = 300
    self.mp = 30

  def walk(self):
    print('walking')

  def rest(self):
    print('resting')

  def run(self):
    print('running')

  def setFire(self):
    print('set fire ball')
    self.mp = self.mp - 10
    
a = Wizard('irocks', 'A')
```

```python
a.mp
```

```python
a.setFire()
```

```python
a.mp
```

### 練習題 

寫一個機器人物件 robot， 機器人需要包含以下功能：

- 使用者在建立物件時，可以給予機器人的年齡 (age) 與性別(gender)

- 為機器人建立函式 getRobotAge: 可以列印出機器人的年齡

- 為機器人建立函式 getRobotGender: 可以列印出機器人的性別



```python
class Robot(object):
  def __init__(self, age, gender):
    self.age = age
    self.gender = gender
  def getRobotAge(self):
    return self.age
  def getRobotGender(self):
    return self.gender

a = Robot(12, 'M')
```

```python
a.age
```

```python
a.gender
```

```python
a.getRobotAge()
```

```python
a.getRobotGender()
```
