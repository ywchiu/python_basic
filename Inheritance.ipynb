### 繼承 (Inheritance)

```python
class Proto(object):
  def __init__(self):
    print('Role Created')
  def walk(self):
    print('walking')
  def whoAmI(self):
    print('I am prototype')

a = Proto()
```

```python
a.walk()
```

```python
a.whoAmI()
```

```python
class Wizard(Proto):
  pass

b = Wizard()
```

```python
b.walk()
```

```python
b.whoAmI()
```

```python
class Wizard(Proto):
  def whoAmI(self):
    print('I am the mighty wizard')
  def setFire(self):
    print('Cast Fire Ball')
```

```python
b = Wizard()
b.whoAmI()
```

```python
b.setFire()
```
