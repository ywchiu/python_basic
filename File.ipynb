## 檔案操作 (File)

### 寫入檔案

```python
# w:write, a:append, b: binary, r:read
f = open('test.txt', 'w')
f.write('Hello World')
f.close()
```

### With 語法

```python
with open('test.txt', 'w') as f:
  f.write('Hello World')
```

### 讀取檔案

```python
with open('test.txt', 'r') as f:
  data = f.read()
print(data)
```

```python
with open('test2.txt', 'w') as f:
  f.write('Hello\nWorld')
```

```python
with open('test2.txt', 'r') as f:
  data = f.read()
print('start printing: ',data)
```

```python
with open('test2.txt', 'r') as f:
  for line in f.readlines():
    print('start printing: ',line)
```

```python
a = '   123 \n\n    '
print(a)
```

```python
print(a.strip())
```

```python
with open('test2.txt', 'r') as f:
  for line in f.readlines():
    print('start printing: ',line.strip())
```

### 計算行數

```python
s = 0
with open('test2.txt', 'r') as f:
  for line in f.readlines():
    s = s + 1
print(s)
```

```python
ary = []
with open('test2.txt', 'r') as f:
  for line in f.readlines():
    ary.append(line)
print(ary)
```

```python
print(len(ary))
```

```python
# pythonic
[line for line in open('test2.txt', 'r')]
```

```python
len([line for line in open('test2.txt', 'r')])
```
