## 函式 (Function)

### 定義函數

```python
def say_hello():
  print('Hello World')
```

```python
say_hello()
```

```python
def say_hello(name):
  print('Hello, ', name)
```

```python
say_hello('David')
```

```python
say_hello('Jack')
```

```python
def add_number(a, b):
  return a + b
```

```python
add_number(3, 5)
```

```python
add_number('Hello', 'World')
```

```python
add_number([1,2,3], [4,5,6])
```

```python
def add_number2(a, b = 3):
  return a * 2 + b
```

```python
add_number2(3,5)
```

```python
add_number2(3)
```

```python
add_number2(3,5)
```

```python
add_number2(b = 3,a = 5)
```

### args

```python
def addall(*args):
  print(args)
  s = 0
  for e in args:
    s = s + e
  return s

addall(1,2,3,4,5)
```

### kwargs

```python
def avg_salary(**kwargs):
  print(kwargs)
  s = 0
  for k,v in kwargs.items():
    s += v
  return s / len(kwargs)

avg_salary(david=50000, marry = 70000, joe = 45000)
```

### 練習題

請撰寫一個函式，當使用者輸入比特幣金額(BTC)至函式時，函式會回傳美金(USD)整數金額

**1BTC = 36,811 USD**



```python
def btc_to_usd(btc):
  btc = float(btc)
  usd = btc * 36811
  usd = int(usd)
  return usd
```

```python
btc = input('請輸入比特幣金額: ')
```

```python
btc_to_usd(btc)
```
