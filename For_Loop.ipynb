### Continue, Break, Pass

```python
for i in range(1,11):
  if i == 3:
    print('find my lucky number')
  print('this number is', i)
```

```python
for i in range(1,11):
  if i == 3:
    print('find my lucky number')
    continue
  print('this number is', i)
```

```python
for i in range(1,11):
  if i == 3:
    print('find my lucky number')
    break
  print('this number is', i)
```

```python
for i in range(1,11):
  if i == 3:
    print('find my lucky number')
    pass
  print('this number is', i)
```

### Enumerate

```python
li = ['David', 'Marry', 'John']
idx = 0 
for p in li:
  print(idx, p)
  idx = idx + 1
```

```python
li = ['David', 'Marry', 'John']

for idx, p in enumerate(li):
  print(idx, p)
```

### 練習題

我們希望建立一個對話機器人, 機器人裡面內建一個資料問答資料庫, 資料庫的問題跟答案如下所示:

- 問題:保險 答案: 以下是我們推薦的險總類
- 問題:貸款 答案: 以下是我們推薦的貸款總類
- 問題:存款 答案: 以下是您的銀行存款餘額

希望你能用這個問答資料庫建立成一個Python 字典, 當使用者輸入問題時, 只要問題中有部分關鍵字對到問答資料庫的問題,電腦就予以回答



```python
question = input('請輸入您的問題:')
print(question)
```

```python
qa = {'保險': '以下是我們推薦的險總類', '貸款': '以下是我們推薦的貸款總類', '存款': '以下是您的銀行存款餘額'}
for k in qa:
  #print(k)
  if k in question:
    print(qa.get(k))
    break
```
