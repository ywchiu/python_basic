## 匿名函式 (Lambda Function)

### Lambda

```python
def square(num):
  ret = num * num
  return ret

square(2)
```

```python
def square(num):
  return num * num

square(3)
```

```python
def square(num): return num * num
square(2)
```

```python
square = lambda num: num * num
square(3)
```

### 偶數

```python
def even(num):
  return num % 2 == 0
even(4)
```

```python
even = lambda num: num % 2 == 0
even(4)
```

### 取第一個元素

```python
def first_ele(s):
  return s[0]

first_ele = lambda s: s[0]
first_ele('Hello')
```

### 字串倒排

```python
def reverse(s):
  return s[::-1]

reverse = lambda s: s[::-1]
reverse('上海自來水來自海上')
```

### 兩個元素相加

```python
def add_number(a,b):
  return a + b

add_number = lambda a,b : a+b
add_number(3,5)
```

### Pandas Map

```python
import pandas
btc =pandas.Series([1,0.1,0.02,0.03])
btc
```

```python
def btc_to_usd(btc):
  return int(btc * 36800)
```

```python
btc.map(btc_to_usd)
```

```python
btc.map(lambda btc: int(btc * 36800))
```
