## 練習題

年化報酬率(%) 的定義為：

$$年化報酬率(\%) = (\ 總報酬率 + 1) ^ {1 / 年數}   - 1$$

如果有兩個投資案：

- 投資案A. 費時10年，總報酬率200%
- 投資案B. 費時5年，總報酬率100%

請問哪個投資案的年化報酬率比較高?




```python
A = (2 + 1) ** (1/10) - 1
A
```

```python
B = (1 + 1) ** (1/5) -1
B
```
